# Face Mask-Detection

```python
import cv2  #importing the open-cv library
import tensorflow as tf
```

```python
cap=cv2.VideoCapture(0)      # creating a object of VideoCapture class 
```

```python
import pickle

model=tf.keras.models.load_model("my_model.hdf5")

def prediction(img):
    
    img_sh=cv2.resize(img,(256,256))
    test_img=img_sh.reshape((1,256,256,3))

    output=model.predict(test_img)
    ot=int(output[0][0])
    
    return ot
```

```python
haar=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):              # creating a fuction which provide the coordinate of the present in image
    cood=haar.detectMultiScale(img)
    return cood
```

```python
import numpy as np
while True:
    ret,frame=cap.read()
    
    cood =detect_face(cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY))
    
    cv2.rectangle(frame,(0,0),(1000,60),(0,0,0),-1)
    cv2.rectangle(frame,(125,70),(550,400),(0,0,0),1)
    
    crop_img = frame[70:400,125:550]
    
    
    
    
    for x,y,w,h in cood:
        
        
        img = frame[y:y+h,x:x+w]
        
        #cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,0),3)
        
        y_pred = prediction(crop_img)
        
        print(y_pred)
        
            
        if y_pred == 1:
            cv2.putText(frame,'NO MASK',org=(250, 40),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1, color=(0,0,255), thickness=2, lineType=cv2.LINE_AA)
        elif y_pred == 0:
            cv2.putText(frame,'MASK',org=(250, 40),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1, color=(0,255,0), thickness=2, lineType=cv2.LINE_AA)
        else:
            cv2.putText(frame,'wait',org=(250, 40),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=1, color=(0,255,0), thickness=2, lineType=cv2.LINE_AA)
    cv2.imshow("Window",frame)
    
    
    
    if cv2.waitKey(1) & 0xFF == ord("x"):
        break

cv2.destroyAllWindows()
```
